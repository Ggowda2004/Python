```python
#data visulization built on top of matplotlib
import seaborn as sns
sns.set_theme(style="darkgrid")
sns.scatterplot(x=[1,2,3,4],y=[1,2,3,4])
```

```python
sns.color_palette()#multiple available ->Set1,2,3
# sns.color_palette("rocket", as_cmap=True) #cmap is used to create a colormap object that can be used for coloring plots based on data values. When as_cmap=True, the function returns a colormap object instead of a list of colors.
```

```python
sns.load_dataset('tips')
#->sns.get_dataset_names()#list of datasets available in seaborn
```

```python
tips_dataset=sns.load_dataset('tips')
sns.scatterplot(tips_dataset, x='total_bill',y='tip',hue='sex', size='size', style='smoker')

#hue is used to add different colors to different data
```

```python
#lineplot
dow_data=sns.load_dataset('dowjones')
sns.lineplot(dow_data, x='Date', y='Price')
```

```python
#more advance approach is using relplot

sns.relplot(tips_dataset, x='total_bill', y='tip', col='day',col_wrap=2, kind='scatter', hue= 'sex', size='size', style='smoker')
#we can change kind according to need -> kind = 'line'
```

```python
sns.displot(tips_dataset, x='total_bill', hue='sex')
#is a more advance version of histplot
```

```python
#for a continous line plot use kdeplot
sns.kdeplot(tips_dataset, x='total_bill', y='tip', fill=True)
```

```python
#more advance approach is using displot
sns.displot(tips_dataset, x='total_bill', y='tip', kind='kde', fill=True, col='day', col_wrap=2)
```

categorical plot

```python
sns.catplot(tips_dataset, x='tip', y='day', kind='strip')
```

```python
sns.boxplot(tips_dataset, x='tip', y='day')#most used
```

```python
#violinplot is also used to show distribution of data but it also shows the probability density of the data at different values.
sns.violinplot(tips_dataset, x='tip', y='day')
```

```python
#countplot is used to show the count of observations in each categorical bin using bars.
sns.countplot(tips_dataset, x='day', hue='sex')
```

joint plot is used to get bot scatter and hsit in a single plot for bot x and y

```python
sns.pairplot(tips_dataset) #pairplot is used to visualize the pairwise relationships between variables in a dataset. It creates a grid of scatter plots for each pair of variables, along with histograms or density plots on the diagonal to show the distribution of each variable. This allows you to quickly identify correlations and patterns in the data.
```

```python
#other most important one is heatmap which is used to visualize the correlation between variables in a dataset.
titanic_data=sns.load_dataset('titanic')
titanic_data.sex= titanic_data.sex.apply(lambda x: 1 if x == 'male' else 0)
titanic_data.alive= titanic_data.alive.apply(lambda x: 1 if x == 'yes' else 0)
titanic_data.alone=titanic_data.alone.apply(lambda x: 1 if x else 0)

titanic_data['class']=titanic_data['class'].apply(lambda x: 1 if x == 'First' else (2 if x == 'Second' else 3))
titanic_data=titanic_data.drop(['who','embark_town','deck', 'embarked', 'adult_male'], axis=1)

titanic_data.corr()#to find correlation between variables
import matplotlib.pyplot as plt
plt.figure(figsize=(12,8))
sns.heatmap(titanic_data.corr(), annot=True, cmap='coolwarm') #annot is used to show the correlation values on the heatmap
```

```python
#last important one is lmplot which is used to visualize the relationship between two variables while also fitting a linear regression model to the data.
sns.lmplot(tips_dataset, x='total_bill', y='tip',hue='sex')#we can also add col and row to create multiple plots based on different categories
```
