```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

'''this line is used to display the plots inline in a Jupyter Notebook. It allows you to see the plots
directly below the code cells that generate them, without needing to call plt.show() explicitly'''
%matplotlib inline

plt.style.use("default") #or we could use 'seaborn' or 'ggplot' etc
```

```python
x=np.array([2025,2026,2027,2028])
y=np.array([4,5,9,7])

plt.plot(x,y)
# if i use only plt.plot(y) -> it will plot y at increment of x with 1 start from 0
# plt.plot(y,x)
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.title('Simple line plot')
plt.show()
#note plt.show() is done autmatically with %matplotlib inline, but using it is a good practice to ensure that the plot is displayed when running the code outside of a Jupyter Notebook.
```

```python
#plot with customizations
y1=np.array([22,3,11,2])
line_style=dict(marker=".",
                markersize=15,
                markerfacecolor="purple",
                markeredgecolor="red",
                linestyle="dotted",
                linewidth=4,)
plt.plot(x, y,color='green', **line_style)
plt.plot(x,y1,color='blue', **line_style)
'''markersize -> ms,  markerfacecolor -> mfc -> use hexadecimal value, markeredgecolor -> mec , linestyle -> ls -> None also,
linewidth=lw, color is line color

if there are multiple lines, we cannot have redundant style code, we create a dict to it, and and unpack 
it with ** when calling plt.plot() for each line
'''
plt.show()
```

```python
#customize the plot with labels and title
plt.title("Customized Line Plot", fontsize=10,
          family="serif", fontweight="bold", color="blue", loc="left")#loc can be left, right, center
plt.xlabel("Year", weight="bold", color="red", fontsize=12)
plt.ylabel("Students", weight="bold", color="red", fontsize=12)
plt.plot(x, y)
plt.tick_params(axis="both",colors="pink")
plt.xticks(x)#to avoid unncessary points to be shown on axis
plt.show()
```

```python
#grind lines
x=[1,2,3,4]
y=[2,3,4,5]
plt.plot(x,y)
plt.grid(axis='x', linewidth=2, color="brown", linestyle="dashed")#y for y
plt.show()
```

```python
#bar graph
menu=["Pizza", "Burger", "Pasta", "Salad"]#list of string dont need to be converted to numpy array
sales=np.array([50, 80, 30, 20])

plt.bar(menu,sales, color='brown')
#for horizontal -> use barh
plt.title("Bar graph for menu")
plt.xlabel("Menu")
plt.ylabel("Quantity sold")

plt.show()
```

```python
#pie chart -> circular

vehicals=["car","bike","auto","truck"]
numbers=np.array([100,270,55,12])

plt.pie(numbers, labels=vehicals, autopct="%1.1f%%",
        explode=[0,0,0.3,0],
        shadow=True)# we can create a list of colors and use it for colors, we can rotate usuing startangle

plt.title("vehical composition")
plt.show() 
```

```python
#scatter graph(+, -, None)

study_hours=[2,3,1,2,5,6,7]
score=[30,44,12,23,70,75,80]

study_hours2=[2,3,1,2,5,6,7]
score2=[32,45,16,24,73,79,80]

plt.scatter(study_hours, score, color="red", label="boy 1")
plt.scatter(study_hours2,score2, color="blue", label="boy 2")
plt.xlabel("hours studied")
plt.ylabel("marks")
plt.legend()
plt.title("study vs hours", color="green")
plt.show()
```

```python
#histograms
# it is a visual representation of the distribution of numerical data, where the data is divided into intervals (bins) and the frequency of data points in each bin is represented by the height of a bar.

scores=np.random.normal(loc=80, scale=10, size=100)#loc is mean, scale is std deviation, size is number of data points

scores=np.clip(scores,0,100)# to limit the values between 0 and 100
plt.hist(scores, bins=10, color="orange", edgecolor="black")#bins is number of intervals, edgecolor is color of bar edges
plt.title("Distribution of Scores")
plt.show()
```

```python
#subplots are multiple plots in a Figure

'''key concepts :
* Figure : top level container(the entire window/page)
* Axes : the actual plot area(where data is drawn) (or a simgle plot/subplot)
* Axis : The number/line objects (x,y etc)
* think of figure as canvas and axes as individual paintings on that canvas'''

figure, axes = plt.subplots(2,2)
axes[0,0].plot(x,y,color="red")
axes[0,0].set_title("Line Plot")
axes[0,1].bar(menu,sales,color="blue")
axes[0,1].set_title("Bar Plot")
axes[1,0].scatter(study_hours,score,color="green")
axes[1,0].set_title("Scatter Plot")
axes[1,1].hist(scores,bins=10,color="orange")
axes[1,1].set_title("Histogram")

plt.tight_layout()# to avoid overlapping of titles and labels
plt.show()
```
