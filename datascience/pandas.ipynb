```python
import pandas as pd
print(pd.__version__)
```

```python
#series in pandas is a one-dimensional labeled array capable of holding any data type (integers, strings, floating point numbers, Python objects, etc.). It can be created from a list, dictionary, or scalar value.
# when creating a pandas Series from a dict, the keys become the index (names) and the values become the data (single values in the Series
data=[10, 20, 30, 40, 50]
series=pd.Series(data)#Series is a constructor in pandas used to create a Series object.
print(series)
data2=[10.1, 20, 30, 40, 50]
series2=pd.Series(data2)
print(series2)
data3=['a', 'b', 'c', 'd', 'e']
series3=pd.Series(data3,index=["first","second","third","fourth","fifth"])
print(series3)
series3.loc["third"]='f'
print(series3.loc["third"])#loc is location based indexing
print(series3.iloc[3])#iloc is integer location based indexing
```

```python
#filtering in series
print(series[series>20])
```

**Dataframes**

```python
#Dataframe in pandas are 2b data(tabular data), with rows and columns. It is a 2-dimensional labeled data structure with columns of potentially different types. It can be created from a dictionary, list of lists, or a 2D array.
data={'Name':['Alice','Bob','Charlie','David'],'City':['New York','Los Angeles','Chicago','Houston']}
df=pd.DataFrame(data)
print(df)
print(df.loc[0])#loc is location based indexing
print(df.iloc[2])#iloc is integer location based indexing
print(df['Name'])#accessing a column

print("-------------------------------------------")
df['Age']=[25,30,35,40]
print(df)

print("-------------------------------------------")
new_row=pd.DataFrame([{'Name':'Eve','City':'Phoenix','Age':28}],index=[4]) #add multiple rows by adding more dictionaries in the list
df=pd.concat([df,new_row])
print(df)
```

Reading a file to use data

```python
df =pd.read_csv("pokemon.csv")#we can read data from csv and json files using pandas, for json files we can use pd.read_json() function
print(df)
#print(df.to_string())
```

using the data from read file

```python
#df is already read above
#selection by column
print(df['Name'])
#print(df['Height','Weight'])this will give an error because we cannot select multiple columns like this, we need to use double square brackets
print(df[['Height','Weight']])

#selection by row
print('-------------------------------------------')
print(df.loc[0])#loc is location based indexing
#we can set the index initially wile reading , to access by column name instead of row number like
# ***pd.read_csv("pokemon.csv",index_col='Name')*** then search by name instead of row number

print(df.loc[0],['Name','Type1'])
print(df.loc[0:5],['Name','Type1'])#this will give us the name and type1 of the first 5 rows

print(df.iloc[0:5:2,0:3])#iloc is integer location based indexing
```

```python
pokemon=input("enter the number of pokemon:")
try:
    print(df.loc[int(pokemon)])
except:
    print("invalid input")
```

Filtering

```python
print(df[df['Height']>=2.0])
print(df[df['Legendary']==True])
print(df[(df['Type1']=='Water') & (df['Type2']=='Poison')])
```

Aggregate functions

```python
#aggreagate in pandas
print(df.mean(numeric_only=True))
print(df.sum(numeric_only=True))#min, max

print("--------------------------------")
print(df['Height'].mean(numeric_only=True))

print("--------------------------------")

group =df.groupby('Type1')
print(group)#not printable
print(group['Height'].mean())
```

Data cleaning

Most important


```python
#drop irrelevant columns
new_df=df.drop(columns=["Legendary"])
print(new_df)
```

```python
#handle missing data
new_df2=df.dropna(subset=['Type2'])#drops rows with missing values in Type2 column
print(new_df2.count())


df2=df.dropna()#drops rows with missing values in any column
print(df2.count())

df3=df.fillna({'Type2':'Unknown'})#fills missing values in Type2 column with 'Unknown'
print(df3.count())
```

```python
#fixing incosnistent data
df['Type1']=df['Type1'].replace({'Grass':'Pollu'})
print(df['Type1'])
```

```python
#standardizing data
df['Name']=df['Name'].str.lower()
print(df['Name'])
```

```python
#fix data types
df['Legendary']=df['Legendary'].astype('bool')
print(df["Legendary"])
```

```python
#remove duplicates
df4=df.drop_duplicates(['Type1'])#drops duplicate rows based on the 'Name' column
print(df4['Type1'])

#to save the cleaned data to a new csv file
df4.to_csv("cleaned_pokemon.csv",index=False)#index = False means removing the added index in the start by pandas
```
